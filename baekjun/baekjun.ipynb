```python
#HanSu Question
def hanSu(n):
    if n < 100:
        return n
    
    elif (n < 1000 or n == 1000):
        count = 0
        for i in range(100, n + 1):    
            ones = i % 10
            tens = (i % 100) // 10
            hundreds = i // 100
            if((hundreds - tens) == (tens - ones)):
                count += 1
        return (count + 99)
    else:
        print('error bound')

num_list = []
for num in range(1, 1001):
    num_list.append(hanSu(num))
print(num_list)
```
